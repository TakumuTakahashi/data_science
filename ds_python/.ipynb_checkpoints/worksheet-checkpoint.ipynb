```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.backends.backend_pdf import PdfPages

x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x)*2
```

```python
from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages('savefig_sample.pdf')

fig, axes = plt.subplots()
axes.plot(x, y1, label='first')
axes.plot(x, y2, label='second')
axes.plot(x, y1+y2, label='first + second')
axes.legend(loc=0)

pdf.savefig(fig)

pdf.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.backends.backend_pdf import PdfPages

x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x)*2
```

```python
pdf = PdfPages('savemultifig_sample.pdf')
for i in range(0, 10):
    #-----グラフ作成------
    fig, axes = plt.subplots()
    axes.plot(x, y1 + x*i)
    axes.set_title('ID:#{}'.format(i))
    #-------------------
    
    #forループのなかで保存
    pdf.savefig(fig)
    
#ループ後に閉じる
pdf.close()
```

```python
def rpstr(num, str='*'):
    if str != '*':
        print('文字列あり---')
        print(str*num)
    else:
        print('文字列なし---')
        print(str*num)
    
    
mozi = input('文字列を入力してください。')
num = int(input('個数を入力してください。'))

rpstr(num, mozi)
    
```
